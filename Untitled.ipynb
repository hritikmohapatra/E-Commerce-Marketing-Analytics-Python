### b. Understanding how many new customers acquired every month

```python
order['order_purchase_timestamp'] = pd.to_datetime(order.order_purchase_timestamp, format = '%m/%d/%Y %H:%M')
```

```python
order['Year_month'] = order.order_purchase_timestamp.apply(lambda x: pd.Timestamp.strftime(x , format = "%Y-%m"))
```

```python
customer_orders = pd.merge(customer , order , how = 'left' , on= 'customer_id')
```

```python
customer_month= customer_orders.loc[:, ['customer_unique_id','Year_month']]
```

```python
customer_month.drop_duplicates('customer_unique_id', inplace=True)
```

```python
cust_acq = customer_month.groupby('Year_month')[['customer_unique_id']].count().reset_index()
cust_acq
```

```python
customer_month.groupby('Year_month')[['customer_unique_id']].count().plot(kind='bar')
```

### 3. Cross-Selling (Which products are selling together) Hint: We need to find which of the top 10 combinations of products are selling together in each transaction. (combination of 2 or 3 buying together)

```python
cross_sell= ord_ord_item_pay_prod_cust[ord_ord_item_pay_prod_cust['order_id'].duplicated(keep=False)]
cross_sell
```

```python
cross_sell['product_id'] = cross_sell['product_id'].astype('str') 
```

```python
cross_sell['Product_Bundle']=cross_sell.groupby('order_id')['product_id'].transform(lambda x: ','.join(x))
cross_sell
```

```python
cross_sell= cross_sell[['order_id','Product_Bundle']].drop_duplicates()
cross_sell
```

```python
from itertools import combinations
from collections import Counter
```

```python
count= Counter()
for row in cross_sell['Product_Bundle']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2))) #counting all the two product bundle
print(count)    
```

```python
#For two product sold together
count.most_common(10)
```

```python
#For three product sold together
count= Counter()
for row in cross_sell['Product_Bundle']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,3))) #counting all the three product bundle
count.most_common(10) 
```
